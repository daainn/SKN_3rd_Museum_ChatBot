```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
from langchain.schema.document import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain_community.utilities import WikipediaAPIWrapper
from langchain.tools import WikipediaQueryRun
import pandas as pd
import re
```

```python
model_name = "Gwangwoon/muse2"

# âœ… í† í¬ë‚˜ì´ì € ë° ëª¨ë¸ ë¡œë“œ
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16
)

# âœ… ë””ë°”ì´ìŠ¤ ì„¤ì •
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print("âœ… MUSE ëª¨ë¸ ë¡œë“œ ì™„ë£Œ!")
```

```python
from langdetect import detect

def select_system_prompt(language: str) -> str:
    if language == "ko":
        return system_prompt
    elif language == "en":
        return system_prompt_eng
    elif language == "ja":
        return system_prompt_japan
    else:
        return system_prompt  # default fallback
```

```python
MAX_HISTORY_MESSAGES = 10  # ì „ì²´ ë©”ì‹œì§€ ê°œìˆ˜ ê¸°ì¤€ (user + assistant í¬í•¨)
chat_history = []

def ask_question(query):
    docs = retriever.get_relevant_documents(query)
    filtered_docs = filter_similar_docs(docs)

    if filtered_docs == "ğŸ”¹ í•´ë‹¹ ì •ë³´ëŠ” ì—†ìŠµë‹ˆë‹¤.":
        print("\nğŸ”¹ í•´ë‹¹ ì •ë³´ëŠ” ì—†ìŠµë‹ˆë‹¤. ìœ„í‚¤í”¼ë””ì•„ ê²€ìƒ‰ì„ ì§„í–‰í•©ë‹ˆë‹¤.")
        return wikipedia_tool.run(query)

    context = "\n".join([doc.page_content for doc in filtered_docs])

    # âœ… ì „ì²´ ë©”ì‹œì§€ êµ¬ì„± (íˆìŠ¤í† ë¦¬ í¬í•¨)
    # messages = [{"role": "system", "content": system_prompt}] + chat_history
    # messages.append({
    #     "role": "user",
    #     "content": f"ì—°ê´€ ì •ë³´:\n{context}\nì§ˆë¬¸: {query}\në‹µë³€:"
    # })

    messages = [{"role": "system", "content": system_prompt}]
    messages += chat_history  # ê³¼ê±° ëŒ€í™” ìœ ì§€
    messages.append({
    "role": "user",
    "content": f"ì—°ê´€ ì •ë³´:\n{context}\n\nì§ˆë¬¸: {query}\në‹µë³€:"})

    # âœ… Chat template ì ìš©
    text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer([text], return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        top_p=0.9,
        temperature=0.3
    )

    output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # âœ… assistant ì´í›„ í…ìŠ¤íŠ¸ë§Œ ì¶”ì¶œ
    if "assistant" in output_text:
        answer = output_text.split("assistant")[-1].strip()
    else:
        answer = output_text.strip()

    # # âœ… ëŒ€í™” íˆìŠ¤í† ë¦¬ì— ëˆ„ì 
    # chat_history.append({
    #     "role": "user",
    #     "content": f"ì—°ê´€ ì •ë³´:\n{context}\nì§ˆë¬¸: {query}\në‹µë³€:"
    # })
    # chat_history.append({
    #     "role": "assistant",
    #     "content": answer
    # })

    chat_history.append({"role": "user", "content": query})
    chat_history.append({"role": "assistant", "content": answer})

    # âœ… íˆìŠ¤í† ë¦¬ ê¸¸ì´ ì œí•œ ì ìš© (ê°€ì¥ ì˜¤ë˜ëœ ê²ƒë¶€í„° ì œê±°)
    if len(chat_history) > MAX_HISTORY_MESSAGES:
        # ê°€ì¥ ì˜¤ë˜ëœ ë©”ì‹œì§€ë¶€í„° ì œê±° â†’ ìµœì‹  MAX_HISTORY_MESSAGESë§Œ ìœ ì§€
        chat_history[:] = chat_history[-MAX_HISTORY_MESSAGES:]\
        
    print("\nğŸ”¹ ì…ë ¥í•œ ì§ˆë¬¸:")
    print(query)
    
    print("\nğŸ”¹ ê²€ìƒ‰ëœ ë¬¸ì„œ:")
    print(context)
    
    print("\nğŸ”¹ AI ì‘ë‹µ:")
    print(answer)
    print("=" * 50)

    return answer
```
